```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
my_path = "D:/data_advanced_ML/"
```

```python

def load_data_new(my_path = my_path):
    list_df = []
    path = my_path + "dunnhumby_Let's-Get-Sort-of-Real-(Full-Part-1-of-9)" + "/" + "dunnhumby_Let's-Get-Sort-of-Real-(Full-Part-1-of-9)/"
    for file in os.listdir(path.replace("\\","/")):
        if file.startswith('tr'):
            list_df.append(pd.read_csv(path + file)[['SHOP_WEEK', 'QUANTITY', 'SPEND', 'PROD_CODE', 'STORE_CODE']])

    df = pd.concat(list_df,axis=0)
    return df

df = load_data_new()
```

```python
df.columns
```

```python
df = df[['SHOP_WEEK', 'QUANTITY', 'SPEND', 'PROD_CODE', 'STORE_CODE']]
```

```python
print('Nombre de magasins uniques:', len(df['STORE_CODE'].unique()))
df = df[df['QUANTITY'] > 0].copy()
#On ne prend que les magasins avec le plus de transactions
nb_stores = 10
top_stores = df['STORE_CODE'].value_counts().head(nb_stores).index.tolist()
df_top = df[df['STORE_CODE'].isin(top_stores)]
```

```python
print('Nombre de produits uniques:', len(df_top['PROD_CODE'].unique()))
nb_products = 50
top_products = df_top['PROD_CODE'].value_counts().head(nb_products).index.tolist()
df_top = df_top[df_top['PROD_CODE'].isin(top_products)]
```

```python
#On prend le prix unitaire par transaction
df_top['UNIT_PRICE'] = df_top['SPEND'] / df_top['QUANTITY']
```

```python
df_final = df_top.pivot_table(
    index=['SHOP_WEEK', 'STORE_CODE'],
    columns='PROD_CODE',
    values='UNIT_PRICE',
    aggfunc='mean'
)
df_final.reset_index(inplace=True)
```

```python
# Nombre de semain sans achat pour chaque produit dans chaque magasin
df_final.isnull().sum() 
```
